# a058. MOD3
> https://zerojudge.tw/ShowProblem?problemid=a058 

相信判斷一個數除以三的餘數是多少，對你來說應該沒有問題。那，如果一次請你判斷很多個數呢嘿嘿？給你 n 個數字，請你輸出 3k、3k+1、3k+2 的數各有幾個?

```python
nn = int(input("輸入正整數 n:"))

a = 0
b = 0
c = 0
for i in range(1, nn, 1):
  if i % 3 == 0:
    a += 1
  elif i % 3 == 1:
    b += 1
  else:
    c += 1
print(f"答案: {a} {b} {c}")
  
```

# d649. 數字三角形
> https://zerojudge.tw/ShowProblem?problemid=d649

```python
hh = int(input("輸入高度: "))

# 靠左版本
for i in range(1, hh + 1):
  for j in range(1, i + 1):
    print('*', end="")
  print(end="\n")
  
print("------------------------------")

# 靠右版本
space_num = hh - 1
for i in range(1, hh + 1):
  print(" " * space_num, end="")
  space_num -= 1
  for j in range(1, i + 1):
    print('*', end="")
  print(end="\n")
```

# 最大公因數
> https://zerojudge.tw/ShowProblem?problemid=a024

```python
# 第 1 版

aa = int(input("輸入正整數: "))
bb = int(input("輸入正整數: "))

tmp = 0
if bb > aa:
  tmp = aa
  aa = bb
  bb = tmp
  
print(f"greater number a : {aa}, less number b : {bb}")

gcd = 1
for nn in range(1, bb + 1):
  if (aa % nn == 0) and (bb % nn == 0):
    gcd = nn

print("GCD is: " + str(gcd))
```

```python
# 第 2 版

aa = int(input("輸入正整數: "))
bb = int(input("輸入正整數: "))

aa, bb = (aa, bb) if aa >= bb else (bb, aa)

print(f"greater number a : {aa}, less number b : {bb}")

for i in range(bb, 0, -1):
  if aa % i == 0 and bb % i == 0:
    print(f"GCD is: {i}")
    break
```

```python
# 第 3 版

aa = int(input("輸入正整數: "))
bb = int(input("輸入正整數: "))

aa, bb = (aa, bb) if aa >= bb else (bb, aa)

print(f"greater number a : {aa}, less number b : {bb}")

gcd = 1
while True:
  remainder = aa % bb
  if remainder == 0:
    gcd = bb
    break
  
  aa = bb
  bb = remainder

print(f"GCD is: {gcd}")
```

```python
# 第 1 版
num = int(input("輸入十進位整數:"))
ans = ""

while num != 0:
  tmp = str(num % 2) # 餘數
  ans = tmp + ans
  num = num // 2
  
print(f"轉2進位: {ans}")
```

# c022. 10783 - Odd Sum
> https://zerojudge.tw/ShowProblem?problemid=c022


給你一個範圍 a 到 b ，請你找出 a 與 b 之間所有奇數的和。
例如：範圍 [3, 9] 中所有奇數的和就是 3 + 5 + 7 + 9 = 24 。

```python
nn = int(input("輸入 n :"))

for i in range(nn):
  aa = int(input("輸入 a :"))
  bb = int(input("輸入 b :"))

  calc_sum = 0
  for nn in range(aa, bb + 1):
    if nn % 2 == 1:
      calc_sum += nn
  print(f"Case {i + 1}: {calc_sum}")
```
